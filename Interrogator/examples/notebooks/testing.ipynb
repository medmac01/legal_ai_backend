## Overview
This notebook demonstrates how to use the Interrogator agent for conducting iterative legal interviews with RAG agents. The Interrogator performs strategic questioning to extract comprehensive legal information and generate detailed reports.

## How to Use

### 1. Installation
First, install the Interrogator package (see cell below).

### 2. Configuration
Configure the agent with basic settings:
- `run_name`: Optional name for the interrogation session

### 3. Initialize and Interrogate
Create an Interrogator instance and call the `interrogation()` method with:
- `userQuery`: The main legal question to be analyzed
- `userContext`: Additional context about the legal scenario
- `userInstructions`: Specific instructions for the interrogation process

### 4. Interrogation Process
The Interrogator conducts a multi-turn conversation:
1. Generates strategic questions based on the user query
2. Gets answers from the RAG agent (Researcher)
3. Refines understanding through follow-up questions
4. Produces a comprehensive final report

### 5. Output
The interrogation returns a structured result containing:
- **Report**: Comprehensive analysis based on the questioning process
- **Conclusion**: Final answer to the original legal question
