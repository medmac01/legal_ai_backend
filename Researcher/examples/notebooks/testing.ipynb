## Overview
This notebook demonstrates how to use the Researcher agent for intelligent information retrieval. The Researcher supports multiple retrieval backends including web search, Wikipedia, vector databases, BM25, hybrid retrieval, and LightRAG graph-based retrieval.

## How to Use

### 1. Installation
First, install the Researcher package (see cell below).

### 2. Configuration
Configure the agent by enabling the desired retrieval backends:
- `enable_web`: Enable web search using Tavily API
- `enable_wikipedia`: Enable Wikipedia article retrieval
- `enable_vectordb`: Enable vector database search (Pinecone/ChromaDB)
- `enable_bm25`: Enable BM25 lexical search
- `enable_hybrid`: Enable hybrid retrieval combining multiple methods
- `enable_lightrag`: Enable LightRAG graph-based retrieval
- `run_name`: Optional name for the search run

### 3. Initialize and Search
Create a Researcher instance with your configuration and call the `search()` method with your query.

### 4. Advanced Features
- **Instructions**: Provide specific instructions to guide the search process
- **Config**: Pass runtime configuration for specific search behaviors
- **Multiple Retrievers**: Combine different retrieval methods for comprehensive results

## Example Usage
The example below shows how to use the Researcher agent with LightRAG for contract analysis tasks.

```python
"""
Simple usage example of the Researcher agent.
"""
from Researcher import Researcher

# Configure the agent
# this config sets the nodes of the graph, while quering you can pick a subset
# of the retrievers to use
config = {
    "enable_web": True,
    "enable_wikipedia": True, # wikipedia retriever returns some times rubbish
    "enable_vectordb": False,  # We don't have a vector db for this example
    "enable_bm25": False,
    "enable_hybrid": True,
    "enable_lightrag": True,
    "run_name": "Example Search"
}

hypothesis = """
Receiving Party shall destroy or return some Confidential Information upon the termination of Agreement.
"""

userQuery = f"Is the following hypothesis ENTAILMENT, CONTRADICTION, or NEUTRAL according to the content of the contract: <hypothesis>{hypothesis}</hypothesis>?"

# Initialize the researcher agent
researcher = Researcher(config)

# Example 1: Direct query
print("\n=== Example 1: Direct Query ===\n")
results = researcher.search(
    query=userQuery,
    instructions="Use light rag tool to search for the answer",
    config={"return_chunks": True}
)
# instructions="Search stored documents in the document database in order to answer the query"
print("Response:")
print(results.get("response", "No response generated"))
```
